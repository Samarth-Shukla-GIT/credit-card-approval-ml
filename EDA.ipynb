```python
# convert to positive days 

df['DAYS_BIRTH'] = df['DAYS_BIRTH'].abs()
df['DAYS_EMPLOYED']=df['DAYS_EMPLOYED'].abs()

# Handle anomalies in DAYS_EMPLOYED

df.loc[df['DAYS_EMPLOYED']>36500, 'DAYS_EMPLOYED'] = 0

# Convert to Years (if required)
df['YEARS_BIRTH'] = (df['DAYS_BIRTH'] /365).astype(int)
df['YEARS_EMPLOYED']=(df['DAYS_EMPLOYED'] /365).astype(int)
```

```python
df
```

```python
# Replace NaN with "Not Determined"
df = df.fillna("Not employeed")

# Verify if nulls remain
print("Null values remaining:", df.isnull().sum().sum())
```

```python
print(df.columns)
```

```python
df.drop(['DAYS_BIRTH','DAYS_EMPLOYED'], axis=1, inplace=True)
```

```python
df.columns
```

```python
# Count how many "Not Determined" in each column
not_determined_counts = (df == "Not Employeed").sum()

# Convert to a DataFrame for a cleaner table
not_determined_table = pd.DataFrame({
    'Column': not_determined_counts.index,
    'Count_Not_Determined': not_determined_counts.values
})

# Show only columns where count > 0
not_determined_table = not_determined_table[not_determined_table['Count_Not_Determined'] > 0]

display(not_determined_table)
```
